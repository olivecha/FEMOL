```python
%autoreload 2
import FEMOL
import numpy as np
import pygmsh
np.set_printoptions(linewidth=200)
np.set_printoptions(suppress=True)
import matplotlib.pyplot as plt
import scipy.sparse
```

## Analytical solution validation to free vibration of orthotropic plates

### 1. FEM Solution

```python
mesh = FEMOL.mesh.rectangle_Q4(1, 1, 15, 15)

problem = FEMOL.FEM_Problem(mesh=mesh, physics='modal', model='plate')

material = FEMOL.materials.OrthotropicMaterial(name='', Ex=100, Ey=10, Es=3.5, vx=0.3, ho = 0.1, rho=1)

layup = FEMOL.laminate.Layup(material=material, plies=[0], symetric=False)

problem.define_materials(material)
problem.define_tensors(layup)

#problem.add_fixed_domain(FEMOL.domains.inside_box([[0, 1]], [[0, 1]]), ddls=[0 ,1])
problem.add_fixed_domain(FEMOL.domains.inside_box([0, 1], [[0, 1]]), ddls=[0, 1, 2, 3, 4])
problem.add_fixed_domain(FEMOL.domains.inside_box([[0, 1]], [0, 1]), ddls=[0, 1, 2, 3, 4])

problem.assemble('K')
problem.assemble('M')

w, v = problem.solve(filtre=0)
```

```python
fig, axs = plt.subplots(2,2, figsize=(8,8))
axs = axs.flatten()
thresh = np.ceil(w[0]*10) / 10
indexes = np.where(w>thresh)[0][:4]
for i, ax in zip(indexes, axs):
    plt.sca(ax)
    mesh.add_displacement(v[i], 6)
    mesh.plot.point_data('Uz', wrapped=False)
    ax.set_title(str(np.around(w[i], 3)) + ' Hz')
plt.show()
```

### 2. Analytical solution

```python
# Modes
k = np.array([3.62, 5.86, 10.45, 13.0])

# D matrix
D = layup.D_mat

# Natural frequencies
w_th = k * np.pi**2 * np.sqrt(D[1,1])
w_th
```
