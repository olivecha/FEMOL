# Cantilever Square Topology Optimization With Real Time Plot

### Topology optimization variables and mesh

```python
nelx = 60 # elements in the x direction
nely = 45 # elements in the y direction
volfrac = 0.1 # Volume fraction of material
penal = 3 # Penalization power

# Creating the mesh
Lx = nelx
Ly = nely

x = np.linspace(0, Lx, nelx)
y = np.linspace(0, Ly, nely)
X, Y = np.meshgrid(x, y)

mesh = RectangleQuadMesh(Lx, Ly, nelx, nely)
mesh.plot()
```

```python
def make_outside_circle(pos_x, pos_y, R):

    def circle(x, y):
        R2 = ((x-pos_x)**2 + (y-pos_y)**2)

        return R2 > ((R) **2)
    
    return circle

def make_outside_box(x1, x2, y1, y2):
    """
    x1 < x2 & y1 < y2
    """

    def box(x, y):
        
        return ~((X > x1) & (X < x2) & (Y > y1) & (Y < y2))
    
    return box

def make_guitar_sides(Lx, Ly):
    angle = np.pi/6
    p = angle/(np.pi/2)
    x1 = 2*Ly/6 + 2*Ly/6 * np.sin(angle)
    y1 = 2*Ly/6 - 2*Ly/6 * np.cos(angle)
    x2 = Lx - Ly/4 - Ly/4 * np.sin(angle)
    y2 = 2*Ly/6 - Ly/4 * np.cos(angle)
    a, b, c, d = create_polynomial(x1, y1, x2, y2, p)
    
    def sides(x, y):
        Y_val = a*X**3 + b*X**2 + c*X + d
        return ~((X > x1) & (X < x2) & (Y > Y_val) & (Y < -Y_val+Ly))

    return sides
    
```

```python
circle1 = make_circle((2*Ly/6), (2*Ly/6), (2*Ly/6))
circle2 = make_circle((2*Ly/6), (4*Ly/6), (2*Ly/6))
circle3 = make_circle((Lx-Ly/4) , 2*Ly/6, Ly/4)
circle4 = make_circle((Lx-Ly/4) , 4*Ly/6, Ly/4)
box1 = make_outside_box(0, Lx, 2*Ly/6, 4*Ly/6)
sides = make_guitar_sides(Lx, Ly)
sound_hole = make_circle(2*Lx/3, Ly/2, Ly/6)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

outside_domain = box1(X, Y) & circle1(X,Y) & circle2(X,Y) & circle3(X,Y) & circle4(X,Y) & sides(X, Y)
rosette = ~sound_hole(X, Y)

ax.scatter(X[outside_domain], Y[outside_domain], c='r')
ax.scatter(X[rosette], Y[rosette], c='b')

ax.set_aspect('equal')
ax.set_axis_off()
```
