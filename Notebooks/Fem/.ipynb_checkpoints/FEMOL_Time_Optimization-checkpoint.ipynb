```python
%autoreload 2
from FEMOL import RectangleQuadMesh, FEM_Problem, Results
from FEMOL_utils import *
import timeit
import numpy as np
np.set_printoptions(linewidth=110)
```

```python
# Create a mesh
Lx = 0.6
Ly = 0.4
nelx = 48
nely = 24

mesh = RectangleQuadMesh(Lx, Ly, nelx, nely)

# Plot the mesh
mesh.plot(node_numbers=False)
```

```python
# choose a material
carbon = general_carbon()
```

```python
# Create FEM problem instance
layup = [0, 0, 0, 0, 0]

# Create the problem
Problem = FEM_Problem(mesh, carbon)

# Get the stress tensor from the layup
Problem.orthotropic_plane_stress_tensor(layup)

# Get the element matrix
Problem.get_plane_stress_Ke()

# Assemble the global matrix
Problem.assemble_plane_stress()
```

```python
# Create a fixed boundary domain

# Fix all the points where x = 0 and y is between [0, 0.1] and [0.3, 0.4]
fixed_x = [0] 
fixed_y = [[0, 0.1], [0.3, 0.4]]

domain = create_domain(fixed_x, fixed_y) # create a domain object

Problem.fix_boundary(domain) # Fix the boundary
Problem.plot()
```

```python
# Define a force and distribute it over a domain
force_domain = [create_domain([0.6], [[0.14, 0.26]])] # domain where the force is applied

force = [[1000000, 0]] # Force vector [Fx, Fy]

Problem.add_forces(force,force_domain) # Add the force on the domains to the problem
Problem.plot()
```

```python
# Solve the problem equation system and get the Result object
Results = Problem.solve()
```

```python
# Mesh displacement plot
Results.mesh_displacement(scale=1)
```

```python
# Total displacement plot
Results.total_displacement(cmap='plasma')
```

```python
# X displacement plot
Results.x_displacement(cmap='plasma')
```
