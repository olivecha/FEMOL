```python
%autoreload 2
from FEMOL import RectangleQuadMesh, FEM_Problem, TOPOPT_Problem
from FEMOL_utils import *
```

#  Orthotropic Cantilever Beams Topology Optimization

```python
volfrac = 0.5 # Volume fraction of material
penal = 3 # Penalization power

# Creating the mesh
# Create a mesh
nelx = 50
nely = 30
Lx = nelx
Ly = nely

mesh = RectangleQuadMesh(Lx, Ly, nelx, nely)

# Plot the mesh
mesh.plot()
```

### Function to create the FEM Problem in the optimization loop

```python
def create_fem_problem_from_layup(layup, mesh, material):
    
    # Creating the FEM Problem
    Problem = FEM_Problem(mesh, material) # initialize the problem
    Problem.orthotropic_plane_stress_tensor(layup) # Define the stiffness tensor
    Problem.get_plane_stress_Ke() # Get the element stiffness matrix (constant for every element)
    
    return Problem
```

### Boundary Conditons and Material

```python
# Define the material
material = T300_N5208()

# Fixed Boundary Conditions
fixed_x = [0] 
fixed_y = [[0, Ly/4], [3*Ly/4, Ly]]

fixed_domain = create_domain(fixed_x, fixed_y) # create a domain object


# Applied Forces
x = [Lx]
y = [[3*Ly/8, 5*Ly/8]]
force_domain = create_domain(x, y)

# Defining the force
force = [[0, -10000]]

# Initiate the density field
X = np.ones(nelx*nely)*volfrac 

plot_Problem = create_fem_problem_from_layup([0], mesh, material)
plot_Problem.assemble_1_material_topopt(X, penal)
plot_Problem.add_fixed_domain(fixed_domain)
plot_Problem.fix_boundary()
plot_Problem.add_forces(force, [force_domain])
plot_Problem.plot()
plt.title("Boudary conditions and applied forces")
plt.show()
```

### Solve for the Layups : [0, 0, 0, 0], [90, 90, 90, 90], [45, -45, -45, 45]

```python
layups = [[0, 0, 0, 0], [90, 90, 90, 90], [45, -45, -45, 45]]
volfrac = 0.5
results = []

for layup in layups:
    Problem = create_fem_problem_from_layup(layup, mesh, material)
    Problem.add_forces(force, [force_domain])
    Problem.add_fixed_domain(fixed_domain)
    TopOpt = TOPOPT_Problem(Problem)
    results.append(TopOpt.solve(converge=0.03))
```

```python
layup_labels = ['[0, 0, 0, 0]', '[90, 90, 90, 90]', '[45, -45, -45, 45]']

plt.figure(figsize=(16,6))
for solution, layup, i in zip(results, layup_labels, range(1,4)):
    plt.subplot(1,3,i)
    solution.plot()
    plt.title(layup)
```
