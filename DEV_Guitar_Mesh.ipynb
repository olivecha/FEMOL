```python
%autoreload 2
from FEMOL.core import *
import FEMOL
import numpy as np
np.set_printoptions(linewidth=200)
import matplotlib.pyplot as plt
import pyvista
import meshio
import meshzoo
import pygmsh
pyvista.set_jupyter_backend('pythreejs')
```

```python
Lx = 1
Ly = 0.2*Lx
lcar = 0.01

with pygmsh.geo.Geometry() as geom:
    # first arc
    p1 = geom.add_point([2*Ly/6, 0], lcar)
    p2 = geom.add_point([2*Ly/6, 2*Ly/6], lcar)
    p3 = geom.add_point([0, 2*Ly/6], lcar)
    arc1 = geom.add_circle_arc(p1, p2, p3)
    
    # Lower bout line
    p4 = geom.add_point([0, 4*Ly/6], lcar)
    l1 = geom.add_line(p3, p4)
    
    # Second arc
    p5 = geom.add_point([2*Ly/6, 4*Ly/6 ], lcar)
    p6 = geom.add_point([2*Ly/6, Ly ], lcar)
    arc2 = geom.add_circle_arc(p4, p5, p6)
    
    c1 = geom.add_circle((Ly/6, Ly/6), Ly/20, lcar, make_surface=False)

    # Line to close the loop
    l2 = geom.add_line(p6, p1)

    loop1 = geom.add_curve_loop([arc1, l1, arc2, l2])
    loop2 = geom.add_curve_loop(c1.curve_loop.curves)
    
    s1 = geom.add_plane_surface(loop1, [loop2])
    
    geom.set_recombined_surfaces([s1])

    
    mesh = geom.generate_mesh()
    
mesh = FEMOL.Mesh(mesh.points, mesh.cells_dict)
mesh.display()
```

```python
def plot_arc(start, center, end):
    # Get the radius
    R = np.sqrt((center[0] - start[0])**2 + (center[1] - start[1])**2)

    # Get the point angles
    theta1 = np.arctan2(start[1] - center[1], start[0] - center[0])
    theta2 = np.arctan2(end[1] - center[1], end[0] - center[0])
    
    while theta2 > theta1:
        theta2 -= np.pi

    x = [start[0]]
    y = [start[1]]
    step = (theta1 - theta2)/60

    while theta1 > theta2:
        theta1 -= step
        x.append(center[0] + R*np.cos(theta1))
        y.append(center[1] + R*np.sin(theta1))

    plt.plot(x, y, color='k')
```
