```python
def analyse_mesh(meshfile, eigvalfile, mode):
    # Load the data
    mesh = FEMOL.mesh.load_vtk(meshfile)
    eigvals = np.load(eigvalfile)
    # Plot the moda of vibration
    fig, ax = plt.subplots()
    plt.sca(ax)
    mesh.plot.point_data('m1_Uz')
    ax.set_title(f'Mode of vibration {mode}')
    # Plot the optimization results
    fig, axs = plt.subplots(5,4, figsize=(16, 20))
    for key, eig, ax in zip(mesh.cell_data.keys(), eigvals, axs.flatten()):
        plt.sca(ax)
        mesh.cell_to_point_data(key)
        mesh.plot.point_data(key, cmap='Greys')
        FEMOL.utils.guitar_outline2(L=1)
        title = f'TOM results {key} eigfreq {int(np.round(eig))} Hz'
        ax.set_title(title)
        
```

```python
lcars = np.linspace(0.1, 0.02, 15)
reference_mesh = FEMOL.mesh.guitar(lcar=0.04)
reference_vectors = [np.load(f'Results/guitar_modes/guitar_mode_T{m}_lcar04.npy') for m in ['11', '21', '12', '31']]

w_values = []
for lcar in lcars:
    print(f'Solving for lcar = {lcar}')
    prob = FEMOL.problems.GuitarModal(mesh_lcar=lcar)
    if 'triangle' not in prob.mesh.contains:
        w, v = prob.solve(mac_find=False)
        mac_vecs = [FEMOL.utils.interpolate_vector(v, reference_mesh, prob.mesh) for v in reference_vectors]
        idxs = [np.argmax([FEMOL.utils.MAC(vi, vref) for vi in v]) for vref in mac_vecs]
        w_values.append(w[idxs])
    else:
        print(f'invalid mesh with lcar = {lcar}')
```

```python
times = np.array([0.16, 0.556, 1.333, 2.183, 2.09, 7.698, 17.44, 27.163, 887.4])
indexing = [True, False, True, False, False, False, True, True, True, False, True, True, True, False, True]
lcar_values = lcars[indexing]
times.shape
```
